```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
titanic=pd.read_csv(r"Titanic.csv")
titanic.head()
```

```python
titanic=pd.get_dummies(titanic)
titanic.head()
```

```python
frequent_itemsets=apriori(titanic,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1,num_itemsets=len(frequent_itemsets))
rules
```
