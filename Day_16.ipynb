```python
import seaborn as sns
import numpy as np
from scipy.stats import zscore
data=sns.load_dataset('iris')
df=data.copy()
z_scores=np.abs(zscore(df.drop('species',axis=1)))

```

```python
threshold = 3
non_outlier = (z_scores < threshold).all(axis=1)
df_no_outlier = df[non_outlier]
```

```python
df_no_outlier
```

```python
threshold=3
outliers=(z_scores>threshold)
outlier_rows=df[outliers.any(axis=1)]
outlier_rows
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset('iris')
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['species']=le.fit_transform(data['species'])
```

```python
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data)
```

```python
clf.predict(data)
```

```python
y_pred_outliers=clf.predict(data)
```

```python
np.where(y_pred_outliers==-1)
```
